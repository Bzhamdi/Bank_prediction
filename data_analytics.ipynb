```python
import csv
import pandas as pd
bank = pd.read_csv('/content/drive/MyDrive/bank-data', sep = ';')
pd.set_option("display.max_columns",50)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 8)
sns.countplot(x = 'age', data = bank)
ax.set_xlabel('Age', fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.set_title('Age Count Distribution', fontsize=15)
sns.despine()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (13, 5))
sns.boxplot(x = 'age', data = bank, orient = 'v', ax = ax1)
ax1.set_xlabel('People Age', fontsize=15)
ax1.set_ylabel('Age', fontsize=15)
ax1.set_title('Age Distribution', fontsize=15)
ax1.tick_params(labelsize=15)

sns.distplot(bank['age'], ax = ax2)
sns.despine(ax = ax2)
ax2.set_xlabel('Age', fontsize=15)
ax2.set_ylabel('Occurence', fontsize=15)
ax2.set_title('Age x Ocucurence', fontsize=15)
ax2.tick_params(labelsize=15)

plt.subplots_adjust(wspace=0.5)
plt.tight_layout() 
```

```python
import matplotlib.pyplot as plt
char = bank.select_dtypes(include='object')
for column in char:
    plt.figure(figsize=(15, 6))
    print(column.title())
    sns.countplot(x = bank[column])
    plt.show()
```

```python
dec = bank.select_dtypes(exclude = 'object')

bank.drop("default", axis=1,inplace=True)
char.drop("default", axis=1)
bank.dropna(inplace=True)
bank.drop(['emp.var.rate', 'nr.employed'], axis = 1, inplace=True)
dec = list(set(dec) - set(['emp.var.rate', 'nr.employed']))
dec
```

```python
i = 1
for column in dec:
    print(column.title())
    plt.subplots(figsize=(16, 16))
    plt.subplot(len(dec)+1, 3, i)
    sns.boxplot(x = bank[column])
    i += 1
    plt.subplot(len(dec)+1, 3, i)
    sns.distplot(x = bank[column])
    i += 1
    plt.subplot(len(dec)+1, 3, i)
    sns.boxplot(x = bank["y"], y = bank[column])
    i += 1
    plt.show()
```
